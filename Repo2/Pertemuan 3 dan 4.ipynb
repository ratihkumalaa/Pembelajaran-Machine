```python
mean = df_noid[['Height']].mean()
stdev = df_noid[['Height']].std()
iso =(df_noid[['Height']] > mean + 3*stdev)
df1 = df_noid[['Stage.Name', 'Weight']].assign(is_outlier=iso)
df1.loc[df1['is_outlier']]
```

```python
q1 = df_noid['Followers'].quantile(0.25)
q3 = df_noid['Followers'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
iso = (df_noid['Followers'] < q1 - 1.5 * iqr) | (df_noid['Followers'] > q3 + 1.5 * iqr)
iseo = (df_noid['Followers'] < q1 - 1.5 - 3*iqr) | (df_noid['Followers'] > q3 + 3* iqr)
```

```python
df1 = df_noid[['Stage.Name', 'Followers']].assign(is_outlier=iso, is_extreme_outlier = iseo)
df1.loc[df1['is_outlier'] | df1['is_extreme_outlier']]
```
